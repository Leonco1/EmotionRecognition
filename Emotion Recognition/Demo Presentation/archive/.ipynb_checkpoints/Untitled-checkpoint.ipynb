```python
Datadirectory="train/"
```

```python
Classes=["0","1","2","3","4","5","6"]
```

```python
for category in Classes:
    path=os.path.join(Datadirectory,category)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB))
        plt.show()
        break
    break
```

```python
img_size=224
new_array=cv2.resize(img_array,(img_size,img_size))
plt.imshow(cv2.cvtColor(new_array,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
new_array.shape
```

```python
training_data=[]

def create_training_Data():
    for category in Classes:
        path=os.path.join(Datadirectory,category)
        class_num=Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img))
                new_arrays=cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
import random
```

```python
random.shuffle(training_data)
```

```python
X=[]
Y=[]
for features,label in training_data:
    X.append(features)
    Y.append(label)
X=np.array(X).reshape(-1,img_size,img_size,3)
#convert to 4d
```

```python
X.shape
```

```python
#normalize
X=X/255.0
```

```python
print(X)
```

```python
Y=np.array(Y)
```

```python
Y.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=tf.keras.applications.MobileNetV2()
```

```python
model.summary()
```

#Transfer learning 





```python
base_input=model.layers[0].input
```

```python
base_output=model.layers[-2].output
```

```python
base_output
```

```python
final_output=layers.Dense(128)(base_output)
final_output=layers.Activation('relu')(final_output)#ACTVATION FUNC
final_output=layers.Dense(64)(final_output)
final_output=layers.Activation('relu')(final_output)
final_output=layers.Dense(7,activation='softmax')(final_output)#IMAME 7 FUNCKII
```

```python
final_output
```

```python
new_model=keras.Model(inputs=base_input,outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
from tensorflow.python.client import device_lib

print(device_lib.list_local_devices())
```

```python
new_model.fit(X,Y,epochs=12)
```

```python
new_model.save("final_model")
```
